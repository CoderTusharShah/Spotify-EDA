#### Checking for column data types

- Removing ' , ' from the numeric data
- Typecasting columns like One Biillion 100 Million , Tracks into int64

```python
spotify.columns
```

```python
spotify.dtypes
```

Before TypeCasting into integer first we have to remove ',' value from the number

```python
spotify['Lead Streams'] = spotify['Lead Streams'].str.replace(',','')
spotify['Feats'] = spotify['Feats'].str.replace(',','')
spotify['Tracks'] = spotify['Tracks'].str.replace(',','')
```

```python
spotify['Lead Streams'] = spotify['Lead Streams'].astype(float)
spotify['Feats'] = spotify['Feats'].astype(float)
spotify['Tracks'] = spotify['Tracks'].astype(float)
spotify['One Billion'] = spotify['One Billion'].astype(float)
spotify['100 Million'] = spotify['100 Million'].astype(float)
```

```python
spotify['Last Updated'] = pd.to_datetime(spotify['Last Updated'])
```

## Exploratory Analysis & Visualization
We will Analyse all the columns & tries to find the relation between them

```python
spotify.columns
```

### Artist
Top 20 Artists by Lead Stream

```python
top20_lead_stream = spotify.sort_values(by=['Lead Streams'] , ascending = False).head(20)
top20_lead_stream
```

```python
plt.bar(top20_lead_stream['Artist Name'],top20_lead_stream['Lead Streams'])
plt.xticks(rotation=75)
plt.xlabel('Artists')
plt.ylabel('Lead Streams')
plt.title('Top 20 Artist by Lead Streams')
```

```python
sns.barplot(x='Lead Streams' , y = 'Artist Name' , data=top20_lead_stream)
```

Top 20 Artist by 100 Million

```python
top20_million = spotify.sort_values(by=['100 Million'] , ascending = False).head(20)
top20_million
```

```python
plt.bar(top20_million['Artist Name'],top20_million['100 Million'])
plt.xticks(rotation=75)
plt.xlabel('Artists')
plt.ylabel('100 Million')
plt.title('Top 20 Artist by 100 Million');
```

```python
sns.barplot(x='100 Million' , y = 'Artist Name' , data=top20_lead_stream)
```

We can see a positive relation between Lead Streams & Million , Artists having more Lead stream tend to have more number of million

```python
top20_billion = spotify.sort_values(by=['One Billion'] , ascending = False).head(20)
top20_billion
```

```python
plt.bar(top20_billion['Artist Name'] , top20_billion['One Billion'])
plt.xticks(rotation = 75)
plt.xlabel('Artist Name')
plt.ylabel('Number of Billions')
plt.title('Top 20 Artisit by Number of Billions')
plt.show()
```

```python
sns.barplot(x='One Billion' , y = 'Artist Name' , data = top20_billion);
```

Now , if we see that plot for Number of Billions we observe that the trends changes a little bits. So , More Lead Streams is not equavalent to more number of billions

### Tracks
Does more number of tracks ensure more 100 Million or 1 Billion?

```python
plt.scatter(spotify.Tracks , spotify['100 Million'])
plt.ylabel('100 Million')
plt.xlabel('Number of Tracks')
plt.title('Relation of Track with 100 Million')
plt.show()
```

```python
sns.scatterplot(x='Tracks',y='100 Million' , data=spotify , s=60)
plt.title('Relation of Track with 100 Million');
```
